# Stage1 Output

```python
fig, ax = plt.subplots(5, 3, figsize = (20, 12))

for i in range(15):
    Actual = "Stage1.Output.Measurement" + str(i) + ".U.Actual"
    Setpoint = "Stage1.Output.Measurement" + str(i) + ".U.Setpoint"
    
    row = i // 3
    col = i % 3
    
    ax[row, col].plot(df1[Actual], "r")
    ax[row, col].plot(df1[Setpoint], "b")

plt.show()
```
